```python
# Evaluation and Final Training for Discovered Architecture

import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from tqdm import tqdm
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import xgboost as xgb
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Import necessary classes from search code
# (Assumes you've run the search code first)
from pathlib import Path

# ==================== CONFIGURATION ====================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_ROOT = "./dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "train")
TEST_DIR = os.path.join(DATA_ROOT, "test")

# ==================== LOAD GENOTYPE ====================
def load_genotype(path="genotype.txt"):
    """Load discovered architecture"""
    with open(path, "r") as f:
        genotype = f.read().strip().split(",")
    return genotype

# ==================== FINAL NETWORK ====================
class FinalNetwork(nn.Module):
    """Final network built from discovered genotype"""
    def __init__(self, genotype, C=32, num_cells=12, num_classes=2):
        super().__init__()
        self.genotype = genotype
        
        # Stem
        self.stem = nn.Sequential(
            nn.Conv2d(3, C, 3, padding=1, bias=False),
            nn.BatchNorm2d(C),
            nn.ReLU()
        )
        
        # Build cells from genotype
        self.cells = nn.ModuleList()
        for i in range(num_cells):
            # Add reduction cell every 1/3 of cells
            is_reduction = i in [num_cells//3, 2*num_cells//3]
            
            if is_reduction:
                cell = ReductionCell(C, C*2, genotype)
                C = C * 2
            else:
                cell = NormalCell(C, genotype)
            
            self.cells.append(cell)
        
        # Classifier
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.classifier = nn.Linear(C, num_classes)
    
    def forward(self, x):
        s = self.stem(x)
        
        for cell in self.cells:
            s = cell(s)
        
        features = self.global_pool(s).view(s.size(0), -1)
        logits = self.classifier(features)
        
        return logits, features

class NormalCell(nn.Module):
    """Normal cell using discovered operations"""
    def __init__(self, C, genotype):
        super().__init__()
        self.ops = nn.ModuleList([
            self._make_op(op_name, C) for op_name in genotype
        ])
    
    def _make_op(self, name, C):
        if name == "sep_conv_3x3":
            return SepConv(C, 3)
        elif name == "sep_conv_5x5":
            return SepConv(C, 5)
        elif name == "dil_conv_3x3":
            return DilConv(C, 3, 2)
        elif name == "avg_pool_3x3":
            return nn.Sequential(
                nn.AvgPool2d(3, stride=1, padding=1),
                nn.BatchNorm2d(C)
            )
        elif name == "max_pool_3x3":
            return nn.Sequential(
                nn.MaxPool2d(3, stride=1, padding=1),
                nn.BatchNorm2d(C)
            )
        elif name == "skip_connect":
            return nn.Identity()
        else:
            # Default to skip connection
            return nn.Identity()
    
    def forward(self, x):
        states = [x]
        for op in self.ops:
            states.append(op(states[-1]))
        
        # Sum all intermediate states
        return sum(states[1:]) / len(self.ops) + x

class ReductionCell(nn.Module):
    """Reduction cell for downsampling"""
    def __init__(self, C_in, C_out, genotype):
        super().__init__()
        
        self.preprocess = nn.Sequential(
            nn.Conv2d(C_in, C_out, 1, bias=False),
            nn.BatchNorm2d(C_out)
        )
        
        self.ops = nn.ModuleList([
            self._make_op(op_name, C_out) for op_name in genotype[:2]  # Use first 2 ops
        ])
        
        self.downsample = nn.MaxPool2d(2, 2)
    
    def _make_op(self, name, C):
        if name == "sep_conv_3x3":
            return SepConv(C, 3)
        elif name == "sep_conv_5x5":
            return SepConv(C, 5)
        else:
            return nn.Sequential(
                nn.Conv2d(C, C, 3, padding=1, bias=False),
                nn.BatchNorm2d(C),
                nn.ReLU()
            )
    
    def forward(self, x):
        x = self.preprocess(x)
        
        states = []
        for op in self.ops:
            states.append(op(x))
        
        out = sum(states) / len(states)
        return self.downsample(out)

# ==================== OPS (same as search) ====================
class SepConv(nn.Module):
    def __init__(self, C, kernel_size):
        super().__init__()
        padding = kernel_size // 2
        self.op = nn.Sequential(
            nn.ReLU(inplace=False),
            nn.Conv2d(C, C, kernel_size=kernel_size, padding=padding, groups=C, bias=False),
            nn.Conv2d(C, C, kernel_size=1, bias=False),
            nn.BatchNorm2d(C),
        )
    
    def forward(self, x):
        return self.op(x)

class DilConv(nn.Module):
    def __init__(self, C, kernel_size, dilation):
        super().__init__()
        padding = dilation * (kernel_size - 1) // 2
        self.op = nn.Sequential(
            nn.ReLU(inplace=False),
            nn.Conv2d(C, C, kernel_size=kernel_size, padding=padding, 
                     dilation=dilation, bias=False),
            nn.BatchNorm2d(C),
        )
    
    def forward(self, x):
        return self.op(x)

# ==================== TRAINING ====================
def train_final_model(genotype, train_loader, test_loader, num_classes, epochs=30):
    """Train the final discovered architecture"""
    
    model = FinalNetwork(genotype, C=32, num_cells=12, num_classes=num_classes).to(device)
    
    print(f"Final model parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    optimizer = torch.optim.SGD(
        model.parameters(),
        lr=0.025,
        momentum=0.9,
        weight_decay=3e-4
    )
    
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=epochs, eta_min=1e-4
    )
    
    best_acc = 0.0
    
    for epoch in range(epochs):
        # Training
        model.train()
        train_loss = 0.0
        train_correct = 0
        train_total = 0
        
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for x, y in pbar:
            x, y = x.to(device), y.to(device)
            
            optimizer.zero_grad()
            logits, _ = model(x)
            loss = F.cross_entropy(logits, y)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = logits.max(1)
            train_total += y.size(0)
            train_correct += predicted.eq(y).sum().item()
            
            pbar.set_postfix({
                'loss': f'{loss.item():.3f}',
                'acc': f'{100.*train_correct/train_total:.2f}%'
            })
        
        scheduler.step()
        
        # Evaluation
        model.eval()
        test_correct = 0
        test_total = 0
        
        with torch.no_grad():
            for x, y in test_loader:
                x, y = x.to(device), y.to(device)
                logits, _ = model(x)
                _, predicted = logits.max(1)
                test_total += y.size(0)
                test_correct += predicted.eq(y).sum().item()
        
        test_acc = 100. * test_correct / test_total
        train_acc = 100. * train_correct / train_total
        
        print(f"\nEpoch {epoch+1}:")
        print(f"  Train Loss: {train_loss/len(train_loader):.4f}, Train Acc: {train_acc:.2f}%")
        print(f"  Test Acc: {test_acc:.2f}%")
        
        # Save best model
        if test_acc > best_acc:
            best_acc = test_acc
            torch.save(model.state_dict(), "best_final_model.pth")
            print(f"  ✓ Saved best model (acc: {best_acc:.2f}%)")
        print()
    
    print(f"Best test accuracy: {best_acc:.2f}%")
    return model

# ==================== FEATURE EXTRACTION & XGBoost ====================
def extract_features(model, loader):
    """Extract features for XGBoost"""
    model.eval()
    features_list = []
    labels_list = []
    
    with torch.no_grad():
        for x, y in tqdm(loader, desc="Extracting features"):
            x = x.to(device)
            _, features = model(x)
            features_list.append(features.cpu().numpy())
            labels_list.append(y.numpy())
    
    X = np.vstack(features_list)
    y = np.concatenate(labels_list)
    
    return X, y

def train_xgboost(X_train, y_train, X_test, y_test):
    """Train XGBoost classifier on extracted features"""
    
    print("\nTraining XGBoost...")
    clf = xgb.XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    )
    
    clf.fit(X_train, y_train)
    
    # Evaluate
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    
    print(f"\nXGBoost Results:")
    print(f"  Accuracy: {acc:.4f}")
    print(f"  Macro F1: {f1:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    # Save model
    joblib.dump(clf, "xgb_final.joblib")
    print("\nSaved XGBoost model: xgb_final.joblib")
    
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix - XGBoost')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
    print("Saved confusion_matrix.png")
    
    return clf

# ==================== VISUALIZATION ====================
def visualize_alphas():
    """Visualize architecture parameters evolution"""
    
    import glob
    alpha_files = sorted(glob.glob("alphas_epoch*.npy"))
    
    if not alpha_files:
        print("No alpha snapshots found")
        return
    
    # Load all alphas
    alphas_history = [np.load(f) for f in alpha_files]
    alphas_history = np.stack(alphas_history)  # (epochs, edges, ops)
    
    # Plot evolution
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    op_names = ["sep_3x3", "sep_5x5", "dil_3x3", "avg_pool", "max_pool", "skip"]
    
    for edge_idx in range(min(6, alphas_history.shape[1])):
        ax = axes[edge_idx]
        
        for op_idx, op_name in enumerate(op_names):
            # Compute softmax probabilities
            probs = []
            for epoch_alphas in alphas_history:
                alpha_row = epoch_alphas[edge_idx]
                exp_alpha = np.exp(alpha_row - np.max(alpha_row))
                prob = exp_alpha / exp_alpha.sum()
                probs.append(prob[op_idx])
            
            ax.plot(probs, label=op_name, marker='o')
        
        ax.set_title(f'Edge {edge_idx}')
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Probability')
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('alpha_evolution.png', dpi=150, bbox_inches='tight')
    print("Saved alpha_evolution.png")

# ==================== MAIN ====================
def main():
    """Main evaluation procedure"""
    
    # Load genotype
    genotype = load_genotype()
    print(f"Loaded genotype: {genotype}")
    
    # Load datasets (assuming same Dataset class from search)
    from torch.utils.data import Dataset
    from torchvision import transforms
    from PIL import Image
    
    class PancreaticDataset(Dataset):
        def __init__(self, root_dir, class_to_idx=None, img_size=224, augment=False):
            self.root = Path(root_dir)
            classes = sorted([d.name for d in self.root.iterdir() 
                             if d.is_dir() and not d.name.startswith('.')])
            
            if class_to_idx is None:
                self.class_to_idx = {c: i for i, c in enumerate(classes)}
            else:
                self.class_to_idx = class_to_idx
            
            self.samples = []
            for cls in classes:
                cls_dir = self.root / cls
                for img_path in cls_dir.glob("*"):
                    if img_path.suffix.lower() in [".png", ".jpg", ".jpeg", ".bmp"]:
                        self.samples.append((str(img_path), self.class_to_idx[cls]))
            
            if augment:
                self.transform = transforms.Compose([
                    transforms.RandomResizedCrop(img_size, scale=(0.8, 1.0)),
                    transforms.RandomHorizontalFlip(),
                    transforms.RandomRotation(15),
                    transforms.ToTensor(),
                    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                ])
            else:
                self.transform = transforms.Compose([
                    transforms.Resize((img_size, img_size)),
                    transforms.ToTensor(),
                    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                ])
        
        def __len__(self):
            return len(self.samples)
        
        def __getitem__(self, idx):
            path, label = self.samples[idx]
            img = Image.open(path).convert("RGB")
            return self.transform(img), label
    
    train_ds = PancreaticDataset(TRAIN_DIR, img_size=224, augment=True)
    test_ds = PancreaticDataset(TEST_DIR, class_to_idx=train_ds.class_to_idx, 
                                img_size=224, augment=False)
    
    train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=0)
    test_loader = DataLoader(test_ds, batch_size=32, shuffle=False, num_workers=0)
    
    # Train final model
    model = train_final_model(genotype, train_loader, test_loader, 
                             num_classes=len(train_ds.class_to_idx), epochs=30)
    
    # Load best model
    model.load_state_dict(torch.load("best_final_model.pth"))
    
    # Extract features
    print("\nExtracting features...")
    X_train, y_train = extract_features(model, train_loader)
    X_test, y_test = extract_features(model, test_loader)
    
    # Save features
    np.save("X_train_feats.npy", X_train)
    np.save("y_train.npy", y_train)
    np.save("X_test_feats.npy", X_test)
    np.save("y_test.npy", y_test)
    
    # Train XGBoost
    clf = train_xgboost(X_train, y_train, X_test, y_test)
    
    # Visualizations
    visualize_alphas()
    
    print("\n" + "="*50)
    print("EVALUATION COMPLETE!")
    print("="*50)

if __name__ == "__main__":
    main()
```
